```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
features = df[['year']]
target = df['per capita income (US$)']
```

```python
lr = LinearRegression()
lr.fit(features, target)
lr.predict([[2020]])
```

```python
lr.predict([[2025]])
```

```python
%matplotlib inline
plt.xlabel("years")
plt.ylabel("Income per capita (US$)")
plt.scatter(features, target, color='green', marker='+')
plt.plot(df.year, lr.predict(features), color='red')
```
