```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
def gradient_descent(x, y, learning_rate=0.01, max_iter=1000, tol=1e-6, visualize=True):
    m_curr = b_curr = 0
    n = len(x)
    prev_cost = float('inf')

    if visualize:
        plt.figure(figsize=(10,6))
        plt.scatter(x, y, color='red', marker='+', linewidth=5)

    for i in range(max_iter):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * np.sum((y - y_predicted) ** 2)

        if i != 0 and cost > prev_cost:
            print(f"Cost increased at iteration {i}, stopping optimization.")
            break
        prev_cost = cost

        md = -(2/n) * np.sum(x * (y - y_predicted))
        bd = -(2/n) * np.sum(y - y_predicted)

        m_curr -= learning_rate * md
        b_curr -= learning_rate * bd

        # Check for convergence
        if abs(md) < tol and abs(bd) < tol:
            print(f"Converged at iteration {i}")
            break

        # Plot only every 50 iterations to avoid overloading
        if visualize and i % 50 == 0:
            plt.plot(x, y_predicted, color='green', alpha=0.3)

    # Final line
    if visualize:
        y_final = m_curr * x + b_curr
        plt.plot(x, y_final, color='blue')
        plt.title("Gradient Descent Progress")
        plt.xlabel("x")
        plt.ylabel("y")
        plt.show()

    return m_curr, b_curr
```

```python
y = np.array([1, 4, 6, 8, 10])
x = np.array([0.5, 2, 3, 4, 5])

m, b = gradient_descent(x, y, learning_rate=0.083)
```
